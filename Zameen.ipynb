```python
import re
```

```python
multiplier = {"crore": 10_000_000, 
              "lakh": 100_000, 
              "arab": 1_000_000_000, 
              "thousand": 1_000}
```

```python
def price_to_pkr(x):
    if pd.isna(x):
        return np.nan
    
    s = str(x).lower()

    m = re.search(r"(\d+(?:\.\d+)?)\s*(crore|lakh|thousand|arab)?", s)

    if not m:
        return np.nan

    value = float(m.group(1))
    unit = (m.group(2) or "").strip()

    return value * multiplier.get(unit, 1)

    
```

```python
zameen_data["price_PKR"] = zameen_data["Price"].apply(price_to_pkr)
```

```python
def area_to_sqft(y):
    if pd.isna(y):
        return np.nan
    
    s = str(y).lower().strip()

    m = re.search(r"(\d+(?:\.\d+)?)", s)

    if not m:
        return np.nan

    value = float(m.group(1))

    if "yd" in s:
        sqft = value * 9
    elif "marla" in s:
        sqft = value * 272.25
    elif "kanal" in s:
        sqft = value * 5445
    else:
        sqft = value

    return f"{int(round(sqft))}"
```

```python
zameen_data["area_sqft"] = zameen_data["Area"].apply(area_to_sqft)
```

```python
num_cols = ['area_sqft', 'price_PKR', 'Bedrooms', 'Bathrooms']

for col in num_cols:
    Q1 = zameen_data[col].quantile(0.25)
    Q3 = zameen_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = zameen_data[(zameen_data[col] < lower) | (zameen_data[col] > upper)]
    print(f"{col}: {len(outliers)} outliers detected")
```

```python
num_cols = ['area_sqft', 'price_PKR', 'Bedrooms', 'Bathrooms']

for col in num_cols:
    Q1 = zameen_data[col].quantile(0.25)
    Q3 = zameen_data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    zameen_data[col] = zameen_data[col].clip(lower, upper)
```

```python
for col in num_cols:
    plt.figure(figsize=(6,3))
    sns.boxplot(x=zameen_data[col])
    plt.title(f'Boxplot after Outlier Treatment - {col}')
    plt.show()
```
